```python
fasta = AlignIO.read("../data/week6/all.aln", "fasta")
```

```python
filenames = !ls ../data/week6/Human
name_dict = dict()

for filename in filenames:
    s = open("../data/week6/Human/" + filename, "r").readlines()[0].strip()[1:]
    splitted = s.split(" ")
    name_dict[splitted[0]] = " ".join(filename.replace(".fasta", "").split()[1:])
```

```python
with open("../data/week6/tree_1/clustalo_default-none-none-fasttree_full/all.aln.final_tree.nwx", "r") as f:
    lines = f.readlines()
    
with open("../data/week6/tree_1/clustalo_default-none-none-fasttree_full/all.fixed.nwx", "w") as f:
    for line in lines:
        for n in name_dict:
            line = line.replace(n, name_dict[n])
        f.write(line)
```

```python
trees = Phylo.parse('../data/week6/tree_1/clustalo_default-none-none-fasttree_full/all.fixed.nwx', 'newick')
fig = plt.figure(figsize=(24, 18))
ax = fig.add_subplot(1, 1, 1)
Phylo.draw(next(trees), axes=ax)
```

Let's assume that average mutation rate is ~1.665e-8 mutations/site/year

Is's taken from (https://www.sciencedirect.com/science/article/pii/S000292970800205X)

```python
avg_mutations = np.mean([tree.distance(f.name) for f in fasta])

print("Estimated time of divergence: ~{}kyr".format(int(avg_mutations / 1.665e-8 / 1000)))
```
